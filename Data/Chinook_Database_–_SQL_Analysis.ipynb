### Invoice-wise Track Details for Invoice ID = 1
This query retrieves detailed information about tracks purchased in a specific invoice,
including track name, media type, quantity, and unit price.

```python
q='''select invoice_id inv_id,tr.track_id track_id,tr.name track_name,il.quantity quantity,
     il.unit_price unit_price,mdt.name media_type_name
     from invoice_line il
     inner join track tr
     on
     il.track_id = tr.track_id
     inner join media_type mdt
     on
     tr.media_type_id = mdt.media_type_id
     '''
```

```python
sql_query(q)
```

### Customer-wise Invoice Track Details for Invoice ID = 1
This query retrieves detailed information for a specific invoice, including
customer name, track details, media type, quantity purchased, and unit price.


```python
q ='''select inv.invoice_id invoice_id,cust.first_name customer_name,tr.track_id track_id,
       tr.name track_name,mdt.name media_type_name,il.quantity quantity,
       il.unit_price unit_price
       from invoice_line il
       inner join invoice inv
       on
       il.invoice_id = inv.invoice_id
       inner join customer cust
       on
       inv.customer_id = cust.customer_id
       inner join track tr
       on
       tr.track_id = il.track_id
       inner join media_type mdt
       on
       tr.media_type_id = mdt.media_type_id
       where il.invoice_id = 1
       '''
```

```python
sql_query(q)
```

### Number of Tracks Sold by Album and Artist
This query calculates the total number of tracks sold for each album,
along with the corresponding artist name.

```python
q='''select count(il.track_id) no_of_tracks_sale,alb.title album_name,
     art.name artist_name
     from invoice_line il
     inner join track tr
     on
     il.track_id = tr.track_id
     inner join album alb
     on
     tr.album_id = alb.album_id
     inner join artist art
     on
     alb.artist_id = art.artist_id
     group by 2'''
```

```python
sql_query(q)
```

### Top 5 Track Sold
This query calculates the Top 5 tracks sold for each album,
along with the corresponding track name.

```python
q='''select sum(il.quantity) track_sale,tr.name track_name,alb.title album_name
     from invoice_line il
     inner join track tr
     on
     il.track_id =tr.track_id
     inner join album alb
     on
     tr.album_id =alb.album_id
     group by 3
     order by 1 desc
     limit 5'''
```

```python
sql_query(q)
```

### Number of Tracks in Each Genre
This query calculates the total number of tracks available in each genre,
helping identify which genres have the largest catalog.

```python
q='''select tr.name track_name,gen.genre_id genre_id,gen.name genre_name,
     count(tr.track_id) track_id
     from track tr
     inner join genre gen
     on
     tr.genre_id = gen.genre_id
     group by genre_name
     order by track_id desc '''
```

```python
sql_query(q)
```

### Top-Spending Customer in Each Country
This query identifies the customer who has spent the most money
in each country, helping understand regional high-value customers.


```python
q='''select cus.first_name||" "||cus.last_name Customer_name,cus.country Country,
     inv.customer_id Customer_id,sum(inv.total) Total_spent
     from customer cus
     inner join invoice inv
     on
     cus.customer_id = inv.customer_id
     group by Country
     order by Total_spent desc
 '''
```

```python
q3 ='''WITH customer_spending AS (
    SELECT
        cus.customer_id AS Best_customer_id,
        cus.first_name || ' ' || cus.last_name AS Customer_name,
        cus.country AS Country,
        SUM(inv.total) AS Total_spent
    FROM customer cus
    INNER JOIN invoice inv
        ON cus.customer_id = inv.customer_id
    GROUP BY
        cus.customer_id, cus.first_name, cus.last_name, cus.country
),
ranked_customers AS (
    SELECT *,
           ROW_NUMBER() OVER (
               PARTITION BY Country
               ORDER BY Total_spent DESC
           ) AS rn
    FROM customer_spending
)
SELECT
    Best_customer_id,
    Customer_name,
    Country,
    Total_spent AS Top_Buyer
FROM ranked_customers
WHERE rn = 1 '''
```

```python
sql_query(q3)
```

### Customer Segmentation Based on Total Spending
This query categorizes customers into spending groups
(Small Spender, Regular, Big Spender) based on their total purchase amount.


```python
q='''Select cus.first_name||" "||cus.last_name customer_name,
     count(inv.invoice_id) No_of_invoices,sum(inv.total) Total_Spent,
     Case
         When sum(inv.total) < 40 Then "Small spender"
         When sum(inv.total) > 100  Then "Big Spender"
         Else "Regular"
     End as customer_category
     from invoice inv
     inner join customer cus
     on
     inv.customer_id = cus.customer_id
     group by cus.customer_id '''
```

```python
sql_query(q)
```

```python
# value_count
q2='''select cus.first_name||" "||last_name customer_name,count(inv.invoice_id) no_of_invoices,
     sum(inv.total) total_spent,
     Case
         When sum(inv.total) < 40 then "Small spender"
         When sum(inv.total) > 100 then "Big spender"
         Else "Regular"
     End as Customer_category
     from invoice inv
     inner join customer cus
     on
     inv.customer_id = cus.customer_id
     group by cus.customer_id
     '''
```
